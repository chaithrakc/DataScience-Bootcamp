# Natural Language Processing Theory

#### Why you may want to use it?

Natural Language Processing serves a lot of use cases when you're dealing with **text** (or) **unstructured text data**.

- Imagine you work for Google News and you want to group news articles by topic

- maybe you work for a legal firm and you need to sift through thousands of pages of legal documents to find the relevant ones.

*This is where natural language processing can help!*

We will want to:
- **compile the documents** in some fashion 
- get features from them so we'll have to **featurize those documents**
- and then **compare their features**.

Let's go through a simple example.

Imagine you have 2 very simple documents having
- "Blue House"
- "Red House"

Featurized based on word count:
- "Blue House" &rarr; (red,blue,house) &rarr; (0,1,1)
- "Red House" &rarr; (red,blue, house) &rarr; (1,0,1)


Simple way to featurize text documents is to featurize based off a word count. So, we transform blue house into a vectorized word counts.

We basically create a vector count of all the possible words throw all the documents in this case they're red, blue and a house and then we just count how many times those words occur in each document.

That means for document "Blue House", we get (0, 1, 1) since Red occurs 0 times blue occurs and house occurs once. Similarly, for "Red House", we get (1, 0, 1) because red occurs once, blue 0 times, house one time. 

*A document represented as a vector of counts is called a **bag of words**.*

Once we have these bags of words vectors you can use co-signed similarity on the vectors to determine similarity of the documents themselves. This is really useful because we're basically treating each document as a vector of features meaning we can perform mathematical operations such as the cosine similarity taking their dot products and then dividing it by the multiplication of their magnitudes or other similarity metrics in order to figure out how similar two text documents are to each other.

"corpus" just stands for the group of all the documents.

**Term frequency** is the importance of the term within that document.

**Inverse document frequency** which is the importance of the term in the corpus.

**D** is the total number of documents and **t** is equal to a number of documents with the term.

The reason we do this is so that we can get not just a word count but also some sort of notation on how important a word is not just relevant to the document but to the entire corpus of all the documents.
