# PANDAS - OPERATIONS
There are lots of operations with pandas that will be really useful to you, but don't fall into any distinct category.
- `col.unique()` : values in a column of a dataframe
- `col.nunique()` : number of unique values in the column
- `col.value_counts()` : similar to Counter python library
- `col.sum()`
- `col.apply()` : applying with custom and built-in functions
- `df.columns` and `df.index` attributes of dataframe
- `df.sort_values(col)`
- `df.isnull()`
- `df.isna()`
- `df.pivot_table()`

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'col1':[1,2,3,4],
    'col2':[444,555,666,444],
    'col3':['abc','def','ghi','xyz']
})
df
```

```python
# finding unique values in column of a dataframe
df['col1'].unique()
```

```python
# finding number of unique values in a dataframe
len(df['col2'].unique())
```

```python
df['col2'].nunique()
```

```python
# Counter 
df['col2'].value_counts()
```

```python
# conditional selection
df[df['col1']>2]
```

```python
# conditional selection with multiple conditions
df[(df['col1']>2) & (df['col2']==444)]
```

```python
# apply method : most powerful tool
def times2(x:int):
    return x*2
```

```python
df['col1'].sum() # using built in function sum() to a column
```

```python
# applying custom function to each data point in a column
df['col1'].apply(times2)
```

```python
# using lamba with apply()
df['col1'].apply(lambda x: x*2)
```

```python
# applying built in function 
df['col3'].apply(len) #length of each string in a column
```

```python
# dropping the column
df.drop('col1', axis=1, inplace=False)
```

```python
# to get column and index names of a dataframe

df.columns  # attribute of dataframe itself
```

```python
df.index #in our case its range index so it retursn rangeIndex object
```

```python
#SORTING & ORDERING DATAFRAME
df.sort_values('col2')
```

```python
# really useful way to find null values in dataframe
df.isnull()
```

```python
df.isna()
```

more examples on pivot table creation using aggregate functions

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
#  aggregating values of col D by taking the sum
df.pivot_table(values='D', index=['A','B'], columns=['C'], aggfunc=np.sum)
```

```python
# We can also fill missing values using the fill_value parameter.

df.pivot_table(values='D', index=['A','B'], columns=['C'], aggfunc=np.sum, fill_value=0)
```

```python
# aggregating by taking the mean across multiple columns.
df.pivot_table(values=['D','E'], index=['A','B'], aggfunc={'D':np.mean, 'E':np.mean}, fill_value=0)
```

```python
# calculate multiple types of aggregations for any given value column
df.pivot_table(values=['D', 'E'], index=['A', 'C'], aggfunc={'D': np.mean,'E': [min, max, np.mean]})
```
