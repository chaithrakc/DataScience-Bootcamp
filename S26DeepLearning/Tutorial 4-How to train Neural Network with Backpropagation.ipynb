# Neural Network Training

$$Loss Function (L) = (y - \hat y)^2$$

Since we had only single record in the above example.

If we have mutiple records, we need to have cost function.

$$ Cost Function (C) = \sum_{i=1}^n (y-\hat y)^2 $$

Main aim is to reduce the cost function with the help of optimizers. optimizers will help us to find derivatives and it will try to change all the weights w.r.t older weights using the formula

$$ w_4(new) = w_4(old) - η \frac{\partial L}{\partial w_4} $$

To minimize the loss, we have many optimizers - Gradient Descent, Stochastic Gradient Descent, etc.,

Learning rate(η) should be very very low so that we reach global minima.

one epoch is forward propagation + backward propagation

forward propagation and backward propagation keeps happening until the Loss function is reduced completely
