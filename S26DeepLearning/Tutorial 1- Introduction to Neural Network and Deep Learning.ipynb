### Perceptron (Simplest type of Neural Network)

### BASIC NEURAL NETWORK ARCHITECTURE

It has input layer, hidden layer, and output layer. Input will be different features identified in the data.



### How does Neural Network Work?

- we are trying to solve some binary classification problem

- Features x1, x2, and x3 are provided to the input layer
- Three features x1,x2, and x3 are getting passed to hidden neuron present in hidden layer1. This neuron does some processing.
- weights w1,w2, and w3: before psasing features to the neurons, there are some weights assigned to each input.
- weights will be passed to hidden neuron. once it is passed to hidden neuron, two types of operations happens
    - step1: summation of weights and inputs $$\sum_{i=1}^n w _ix_i$$
    - step2: summation is passed through activation function

This is called **Forward Propogation**.


When output from hidden layer is passed to output layer, again their will be weights and bias associated to it.

This is called **Forward Network**.

### Activation Function

##### Sigmoid Activation Function

    summation of inputs and weights (y) gets transformed to a value between 0 and 1. 
    
    if y < 0.5, we are going to consider as 0
    if y > 0.5, we are going to consider as 1
    
    output of 1 indicates that the neuron is activated -> neuron is actually transferring the signal -> helping you to classify the final output.
    
    output of 0 indicates that the neuron is deactivated

##### RELU Activation Function 

$$ max(y,0) $$

    if y is -ve value, max(-ve,0) = 0, neuron is deactivated for all negative values of y
    if y is +ve value, max(+ve, 0) = +ve number. y gets transformed to that particular number if y is +ve.

##### Used for regression problems in the hidden layers, but in  final output layer, we should use sigmoid since it transforms the output to 0 and 1
