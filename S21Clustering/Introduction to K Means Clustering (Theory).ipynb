### INTRODUCTION TO K MEANS CLUSTERING

K Means Clustering is an unsupervised learning algorithm that will attempt to group similar clusters together in the data.

Typical clustering problems look like:
- clustering similar documents
- cluster customers based on features
- market segmentation
- identify similar physical groups


Overall goal is to divide data into distinct groups such that observations within each group are similar.


#### K Means Clustering
- choose a number of clusters K (K value is nothing but centroids).
- randomly assign each point to a cluster (Initialize centroids randomly)
- until clusters stop changing, repeat the following:
    - Similarity Measure: Find the points that are near to the centroids using **Euclidean/Manhattan Distance**
    - Finding new centroid: for each cluster, compute new centroid by taking the mean of points in that cluster
    - update the centroid: assign each data point to the cluster for which the centroid is the closest



### CHOOSING K VALUE

We have to decide how many clusters we expect in the data. The problem of selecting K isn't actually that simple.

- There is no easy answer for choosing a best case value.

- However, one way to try to do it is using something known as the elbow method

#### ELBOW METHOD
- Compute SSE for some values of K (for example 2,4,6,8 etc.,)
- SSE is defined as the sum of squared distance between each member of the cluster and its centroid
- If we plot K against SSE, we will see that the error decreases as K gets larger; this is because when the number of clusters increases, they should be smaller, so distortion is also smaller
- The idea of the elbow method is to choose the K at which the SSE decreases abruptly
- This produces an "elbow effect" in the graph
