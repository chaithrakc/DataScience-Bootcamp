# Hierarchical Clustering

This is unsupervised machine learning technique. 

Produces a set of nested clusters organized as a hierarchical tree. Can be visualized as a dendrogram - A tree like diagram that records the sequences of merges or splits.



- Finding most nearest two points: Algorithm will start combing nearest two points as one cluster. Simultaneously dendogram is plotted. dendogram is distance between points/clusters vs points. will start with shortest distance till longest distance
- bottom up approach

### Hierarchial clustering takes more time than K-Means clustering. Dataset is small, hierarchial clustering cab be used. Dataset is large, K-Means can be used.

## VALIDATE CLUSTERING MODELS (Silhouette Score)

1. computing average distance a(i)
2. computing b(i): distance between each point from cluster2 to a point in cluster1. then take the average of all the distances.
3. if we have a really goog model, b(i) >> a(i)
4. silhouette score varies from -1 to 1. closer to 1, model is good
