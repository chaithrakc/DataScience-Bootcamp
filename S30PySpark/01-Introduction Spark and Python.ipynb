```python
from pyspark import SparkContext

sc = SparkContext()
```

```python
%%writefile example.txt
Chaithra Kopparam Cheluvaiah
Graduate student at iSchool
MS in Applied Data Science
Syracuse, New York
```

```python
example_file = sc.textFile('example.txt') 
# 'example_file' object is the RDD
# sc is spark context that connects to spark cluster
```

RDD have actions that returns values and transformation returns pointers to another RDD

**ACTIONS**

```python
example_file.count()
```

```python
example_file.first()
```

**TRANSFORMATIONS**

```python
search_key = example_file.filter(lambda line: 'Data' in line)
```

```python
search_key
```

```python
search_key.collect()
```

Transformations are not executed until we call the actions
