```python
import matplotlib.pyplot as plt
# allows to see the plots inside the notebook otherwise plt.show() has to be added at the end of all the commands for plotting
```

```python
import numpy as np

x = np.linspace(1,5,5)
y = x ** 2

print(x)
print(y)
```

```python
# functional method of plotting
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#creating multiple plots on the same canvas

# args - num of rows, num of cols, plot number we are working on
plt.subplot(1,2,1)
plt.plot(x,y,'r')  # third argument is color

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented API method of plotting - instantiate figure objects and called methods and attributes from that object
# this is better way of using matplotlib

fig = plt.figure() # Figure object gets created- just creates an empty canvas

# we can think of this as imaginary blank canvas
```
